```python
#meter mi pais 
berlin_data= pd.read_csv("Berlin_Alemania.csv")
pd.options.display.max_rows = 10
pd.options.display.max_columns = 10
berlin_data.head()
```

***PARTE DOS***

```python
#review_scores_value Y super host
filtroA=berlin_data[(berlin_data["review_scores_value"] >= 3) & (berlin_data["host_is_superhost"]=="t")]
print(filtroA)
filtroA.to_csv("scores_SH.csv")
```

```python
#propiedades que no tienen un baño
filtroB = berlin_data[berlin_data["bathrooms"] != 1] 
print(filtroB)
filtroB.to_csv("bathromms.csv")
```

```python
#precio entre 100 y 200
filtroC = berlin_data.loc[(berlin_data["price"] >= "$100") & (berlin_data["price"] <= "$200")]  
filtroC
filtroC.to_csv("entre100_200.csv")
```

```python
#instant_bookable y host_indetity_verified 
filtroD=berlin_data[(berlin_data["host_identity_verified"] =="f") & (berlin_data["instant_bookable"]=="t")]
print(filtroD)
filtroD.to_csv("bookable_indetity.csv")
```

```python
#hotel room o  Private room 
filtroE = berlin_data[(berlin_data["room_type"] =="Hotel room") | (berlin_data["room_type"]=="Private room")]
print(filtroE)
filtroE.to_csv("Hotelroom_private.csv")
```

```python
#mas de 4 accomadtes
filtroF = berlin_data[berlin_data["accommodates"]>4] 
print(filtroF)
filtroF.to_csv("accomodates.csv")
```
